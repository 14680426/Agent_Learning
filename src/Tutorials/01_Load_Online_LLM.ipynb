```python
import openai
import json
from dotenv import load_dotenv
import os
load_dotenv(override=True)
BingXing_API_KEY = os.getenv('BINGXING_API_KEY')
BingXing_BASE_URL = os.getenv('BINGXING_BASE_URL')

client = openai.OpenAI(
    api_key=BingXing_API_KEY,  
    base_url=BingXing_BASE_URL
)

response = client.chat.completions.create(
    model="Qwen3-32B",
    messages=[{"role": "user", "content": "ä½ å¥½ä½ æ˜¯è°ï¼Ÿ"}],
    extra_body={
        "enable_thinking": False  # ğŸ‘ˆ å…³é”®ï¼šå…³é—­ thinking åŠŸèƒ½
    }
)
# ç»“æ„åŒ–æ‰“å°æ•´ä¸ª response
print("response æ•°æ®ç±»å‹ï¼š",type(response))
print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))
print("å›ç­”å†…å®¹ï¼š",response.choices[0].message.content)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

llm = ChatOpenAI(
    model="Qwen3-32B",
    api_key=BingXing_API_KEY,  
    base_url=BingXing_BASE_URL,
    default_headers={"Accept": "application/json"},
)

response = llm.invoke(
    [HumanMessage(content="ä½ å¥½ä½ æ˜¯è°ï¼Ÿ")],
    extra_body={"enable_thinking": False}
)
# ç»“æ„åŒ–æ‰“å°æ•´ä¸ª response
print("response æ•°æ®ç±»å‹ï¼š",type(response))
print(json.dumps(response.model_dump(), indent=2, ensure_ascii=False))
print(response.content)
```

# æ¨¡å‹çš„è¾“å‡ºæ ¼å¼åŒ–

æ¨¡å‹çš„è¾“å‡ºæ ¼å¼åŒ–æ˜¯æŒ‡é™åˆ¶å¤§å‹è¯­è¨€æ¨¡å‹çš„è¾“å‡ºä¸ºç‰¹å®šæ ¼å¼æˆ–ç»“æ„çš„æŠ€æœ¯ï¼Œä¹Ÿç§°ä¸ºç»“æ„åŒ–è¾“å‡ºã€‚è™½ç„¶å¤§æ¨¡å‹èƒ½å¤Ÿç”Ÿæˆä»»æ„æ–‡æœ¬æ¥é€‚åº”å¹¿æ³›çš„è¾“å…¥åœºæ™¯ï¼Œä½†åœ¨æŸäº›ç‰¹å®šåº”ç”¨åœºæ™¯ä¸‹ï¼Œéœ€è¦æ¨¡å‹æŒ‰ç…§é¢„å®šä¹‰çš„æ ¼å¼è¿›è¡Œè¾“å‡ºï¼Œè¿™ä½¿å¾—åç»­çš„æ•°æ®å¤„ç†å’Œåº”ç”¨é›†æˆæ›´åŠ é«˜æ•ˆå¯é ã€‚

è¿™ç§æŠ€æœ¯çš„ä¸»è¦ç”¨é€”åŒ…æ‹¬ï¼š
1. **æ„å›¾è¯†åˆ«**ï¼šè®©æ¨¡å‹æŒ‰ç…§å›ºå®šçš„JSONæ ¼å¼è¾“å‡ºç”¨æˆ·æ„å›¾åˆ†ç±»ç»“æœï¼Œä¾¿äºç¨‹åºè‡ªåŠ¨åŒ–å¤„ç†
2. **æ•°æ®æå–**ï¼šä»éç»“æ„åŒ–æ–‡æœ¬ä¸­æå–ç»“æ„åŒ–ä¿¡æ¯ï¼Œå¦‚å®ä½“è¯†åˆ«ã€å…³ç³»æŠ½å–ç­‰
3. **APIé›†æˆ**ï¼šç¡®ä¿æ¨¡å‹è¾“å‡ºèƒ½å¤Ÿç›´æ¥ä½œä¸ºä¸‹æ¸¸ç³»ç»Ÿçš„è¾“å…¥ï¼Œå‡å°‘æ•°æ®è½¬æ¢æˆæœ¬
4. **æ•°æ®åº“å­˜å‚¨**ï¼šå½“éœ€è¦å°†æ¨¡å‹è¾“å‡ºå­˜å‚¨åˆ°å…³ç³»æ•°æ®åº“æ—¶ï¼Œç»“æ„åŒ–çš„è¾“å‡ºæ ¼å¼å¤§å¤§ç®€åŒ–äº†æ•°æ®å…¥åº“æµç¨‹

é€šè¿‡LangChainç­‰æ¡†æ¶æä¾›çš„`.with_structured_output()`æ–¹æ³•ï¼Œå¼€å‘è€…å¯ä»¥è½»æ¾åœ°å°†æ¨¡å‹è¾“å‡ºçº¦æŸä¸ºç‰¹å®šçš„å­—å…¸æˆ–Pydanticå¯¹è±¡æ ¼å¼ï¼Œæ—¢ä¿è¯äº†è¾“å‡ºçš„è§„èŒƒæ€§ï¼Œåˆç»´æŒäº†å¼€å‘çš„ä¾¿æ·æ€§ã€‚

```python
from typing import List
from pydantic import BaseModel, Field

llm = ChatOpenAI(
    model="DeepSeek-R1-0528", 
    api_key=BingXing_API_KEY,
    base_url=BingXing_BASE_URL,
)
# å®šä¹‰ç”µå½±ä¿¡æ¯çš„ Pydantic æ¨¡å‹
class Movie(BaseModel):
    """ç”µå½±è¯¦æƒ…ã€‚"""
    title: str = Field(..., description="ç”µå½±æ ‡é¢˜")
    year: int = Field(..., description="ç”µå½±å‘è¡Œå¹´ä»½")
    director: str = Field(..., description="ç”µå½±å¯¼æ¼”")
    rating: float = Field(..., description="ç”µå½±è¯„åˆ†ï¼ˆæ»¡åˆ†10åˆ†ï¼‰")
    cast: List[str] = Field(default_factory=list, description="ä¸»è¦æ¼”å‘˜åˆ—è¡¨")

# ä½¿ç”¨ with_structured_output åˆ›å»ºç»“æ„åŒ–è¾“å‡ºæ¨¡å‹
model_with_structure = llm.with_structured_output(Movie)

# è°ƒç”¨æ¨¡å‹
response = model_with_structure.invoke("æä¾›ç”µå½±ã€Šç›—æ¢¦ç©ºé—´ã€‹çš„è¯¦ç»†ä¿¡æ¯")

# æ‰“å°ç»“æœ
print(response)
```

```python
# ä½¿ç”¨ with_structured_output åˆ›å»ºç»“æ„åŒ–è¾“å‡ºæ¨¡å‹
model_with_structure = llm.with_structured_output(Movie, include_raw=True)#æ‰“å°åŸå§‹ä¿¡æ¯

# è°ƒç”¨æ¨¡å‹
response = model_with_structure.invoke("æä¾›ç”µå½±ã€Šç›—æ¢¦ç©ºé—´ã€‹çš„è¯¦ç»†ä¿¡æ¯")

# æ‰“å°ç»“æœ
print(response)
```

**æœ‰äº›æ¨¡å‹ä¸æ”¯æŒä½¿ç”¨with_structured_outputæ–¹æ³•ï¼Œè¿™é‡Œä¹Ÿæ”¯æŒè‡ªå®šä¹‰çš„æ–¹å¼å®ç°ï¼Œä¸»è¦æ˜¯é€šè¿‡æ„å»ºPromptæ–¹å¼ï¼Œç„¶åè°ƒç”¨SimpleJsonOutputParserè§£æ**

```python
from langchain_core.output_parsers import SimpleJsonOutputParser
from langchain_core.prompts import ChatPromptTemplate
# åˆ›å»ºèŠå¤©æç¤ºæ¨¡æ¿ï¼Œè¦æ±‚æ¨¡å‹ä»¥ç‰¹å®šæ ¼å¼å›ç­”é—®é¢˜
prompt = ChatPromptTemplate.from_template(
"å°½ä½ æ‰€èƒ½å›ç­”ç”¨æˆ·çš„é—®é¢˜ã€‚" # åŸºæœ¬æŒ‡ä»¤
'ä½ å¿…é¡»å§‹ç»ˆè¾“å‡ºä¸€ä¸ªåŒ…å«"title","year","director","rating"é”®çš„JSONå¯¹è±¡ã€‚'
"{question}" # ç”¨æˆ·é—®é¢˜å ä½ç¬¦
)

chain = prompt | llm | SimpleJsonOutputParser()
resp = chain.invoke({"question": "æä¾›ç”µå½±ã€Šç›—æ¢¦ç©ºé—´ã€‹çš„è¯¦ç»†ä¿¡æ¯ï¼Ÿ"})
print(resp)
```
